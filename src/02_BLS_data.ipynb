```python
# Import necessary libraries
import config  # Ensure config.py is in the same directory
import json
import requests
import pandas as pd

# The URL for BLS API v2
url = 'https://api.bls.gov/publicAPI/v2/timeseries/data/'

# Include the API key from the config file
key = f'?registrationkey={config.bls_key}'

# Series stored as a dictionary
series_dict = {
    'LNS14000003': 'White',
    'LNS14000006': 'Black',
    'LNS14000009': 'Hispanic'
}

# Start year and end year
dates = ('2008', '2017')

# Specify JSON as the content type to return
headers = {'Content-type': 'application/json'}

# Prepare data for the POST request
data = json.dumps({
    "seriesid": list(series_dict.keys()),
    "startyear": dates[0],
    "endyear": dates[1]
})

# Make the POST request
response = requests.post(
    f'{url}{key}',
    headers=headers,
    data=data
)

# Check if the request was successful
if response.status_code == 200:
    results = response.json()['Results']['series']
else:
    raise Exception(f"API request failed with status code {response.status_code}: {response.text}")

# Process the results
# Extract the date index from the first series
date_list = [f"{i['year']}-{i['period'][1:]}-01" for i in results[0]['data']]

# Initialize an empty DataFrame
df = pd.DataFrame()

# Loop through each series and build the DataFrame
for s in results:
    series_name = series_dict.get(s['seriesID'], s['seriesID'])
    df[series_name] = pd.Series(
        index=pd.to_datetime(date_list),
        data=[float(i['value']) for i in s['data']]
    ).iloc[::-1]  # Reverse the order to make it chronological

# Show the last 5 rows of the DataFrame
print(df.tail())
```
